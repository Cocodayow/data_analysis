# __Background & Prior Work__

>Humans are born to eat well, and finding a good place to dine is a recurring theme in our lives. In an era where collective intelligence on the internet prevails, reading others’ opinions about food is just as important as actually eating the food. According to a survey about dinner decision making, about 94% of US citizens reported that online customer reviews impacted on where they ate out in some way [^Tripadviser, 2018]. Other social science studies also confirmed this phenomena by stating that over half of the dining-out demographic rely on those digital reviews to decide their dining place [^RestLabs] ,[^BloomIntelligence, 2017]. While referring to netizens’ restaurant evaluations indicates a new trend of dietary culture among the customers, this tendency also signifies a potential shift in business strategy among the restaurants. For example, several research studies have noticed positive impacts on a restaurant’s profitability created by positive online rating reviews [^RestLabs, 2023], [^Arevalo, 2017] ,[^Weisskopf & Masset, 2018]. On the flip side, influence generated by negative reviews could seriously countervail with the positive effect [^Luca, 2016]. From this perspective, the key to a successful restaurant business lies in getting more positive rating reviews and less negative ones.

>So what can a restaurant do to improve their online customer reviews? Exploring the factors which correlate with positive reviews is crucial to answering this question. Whereas most existing literature focuses on the environmental aspects like quality of food and service in a restaurant [^Rajput & Gahfoor, 2020], [^Gupta et al., 2022], [^Liu, 2018], few of those studies look at an alternative angle – the personal aspects of the customers who write reviews about the restaurants. In other words, maybe not only how the restaurants serve their customers matters, but what kind of customers go to the restaurants also matters.

>In fact, the idea of this speculation stems from several scientific research. In an experiment conducted by Ketelaar et. al centering how personal attributes of customer affect purchase intention, the researchers noted that experienced customers of a kind of product are less likely to be discouraged by negative reviews to buy the product [^Ketelaar et al., 2015]. Given that this result implies a difference in how people receive online reviews based on personal factors – in this case, the expertise of shopping – it’s possible that such factors also play a role in how people produce online reviews. Similar concepts present in another study which investigates the relationship between self-esteem and consumption behavior: buyers with low self-esteem are significantly inclined to purchase inferior products [^Stuppy et al., 2020]. This finding, if it could apply to the restaurant scenario, might suggest that customers who perceive themselves as less worthy have a higher chance of writing negative reviews. Further, other personal factors such as social networks could also shape the undertone of user reviews. For instance, Daehwan et al. pointed out that travelers who travel with family, lovers, or friends tend to leave more positive reviewing messages about hotel services than those who are by themselves [^Daewan et al., 2017]. Since higher satisfaction of social relationships co-occurs with higher satisfaction of external conditions, this discovery could hint that customers who have a supportive inner social circle are more likely to positively rate public places like restaurants. Consequently, once we single out the impact of personal attributes on customer ratings about restaurants, we can promote those businesses to grow. Specifically, through adjusting marketing strategies which consider that different groups of customers may have distinctive needs, those restaurants can better tailor their service to existing customers and expand their customer scale in the future.

## Testing our Hypothesis More Scientifically

As the above bar charts showed, some visualization seemed to disprove our hypothesis (like people who have kids give lower rating scores than people who are independent). In order to see if our hypothesis should be statistically rejected, we decide to run a t-test on our user data.

  - select the data of people who are married, bring their family/friends to dine, have kids. Store it into variable __hyp_1__.
  - select the people who are single/widowed, dine solitarily, and are independent/have dependents at home. Store it into variable __hyp_01__.

  - select the people who smoke and drink casualy/socialy. Store it into variable __hyp_02__.

### Discussion

While there are some significant relationships between personal factors and their total rating scores, we may identify more notable findings had we diligently addressed limitations of the research process.

- uneven distribution of values within each variables
  - Despite a immensive amount of data (~1000), the distribution of personal traits within individual variables is highly uneven. For example, while 912 costumers were single, only 64 people were married. This sharp difference of population undeniabely affects the average rating scores of each demographic, which could overestimate or underestimate the comparison among different values.

- ambiguous meaning of values
  - perhaps the most problematic aspect of data analysis is interpreting the meaning of certain values. Particularly, we found it difficult to compe up with conclusive implications of the three different values 'kids', 'independent', and 'dependent' contained in the "hijos" variable. Do those things mean 'having kids', 'not having kids', and 'I am a dependent to other people'? Or do they mean 'having small kids', 'having independent kids', and 'having kids who need special care'? Since we interpreted it in one way (not necessarily the right way), we could misread the correlation relationship between those values and the overall rating scores.

- limited range of values
  - Lastly, there could be bias in the unprocessed data in the first place. Because the allowed number of options within each variable does not exceed 3, the resulting categorization of personal traits may not accurately reflect the diversity of the population. For instance, there were 'solitary', 'family', and 'friends' in the ambience variable, but these three concepts do not cover all types of ambience. What about people who dine with their unmarried romantic partner? This incomplete inclusivenesss could also weaken the significance of our finding.
